## import semua library packages yang kita butuhkan untuk visualisasi yaitu pandas,matplotlib,dan seaborn,jangan lupa memanggil datasetnya dan tampilkan data frame tersebut.


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('weatherAUS.csv')
df.head()
```

## kita set index nya,untuk itu saya akan gunakan 'Date' atau tanggal sebagai indexnya.


```python
df.set_index('Date', inplace=True)
df.head()
```

## selanjutnya saya akan membuat variabel  location dan juga menjadikan group untuk pemvisualisasian data

```python
pre= df['Location'].value_counts()
pre
```

```python
pre_group= df.groupby('Location').sum()
pre_group.head(20)
```

```python
#pandas
#Grafik batang
pre_group['MinTemp'].plot.bar()
plt.title("Frekuensi minimal Temperatur setiap lokasi")
plt.xlabel('lokasi')
plt.ylabel('frekuensi')
plt.show()
```

## Kesimpulan grafik batang ini bahwa lokasi 'Darwin' menjadi tempat dengan suhu rendah paling banyak dengan total mencapai lebih dari 74000.

```python
#matplotlib
plt.hist(df['MinTemp'])
plt.title('Prediksi suhu dalam Histogram')
plt.xticks(rotation=90)
plt.xlabel('Prediksi')
plt.ylabel('Frekuensi')
plt.show()
```

## dari histogram bentuk pendistribusian data yaitu suhu 10 derajat menjadi suhu yang frekuensi nya paling tinggi mencapai lebih dari 35000 berdasarkan dari -10 sampai 30 derajat

```python
#piechart
plt.figure(figsize=(10,12))
pre_group['MinTemp'].plot.pie(autopct='%.2f')

```

## dari pie chart di atas di simpulkan bahwa lokasi di 'Darwin' menjadi tempat dengan prediksi suhu rendah paling tinggi mencapai 4.2% sama seperti grafik batang di atas namun dalam versi persentase

```python
total_corr = df.loc[:, "Location":'RainTomorrow'].corr()
total_corr
```

```python

sns.heatmap(
    data=total_corr,
    annot=True
)
plt.gcf().set_size_inches(10, 10)
plt.show ()
```
