## langkah pertama yaitu kita import packages pandas dan prettytable buat membuat dataframe dan juga membuat table agar lebih rapi,selanjutnya kita read datasetnya menggunakan pd.read_csv dan nama file dataset.kemudian tampilkan rata-rata, standar deviasi, nilai minimum, nilai maksimum, Q1, Q2, dan Q3 dari attribute bertipe angka.

```python
df=pd.read_csv('weatherAUS.csv')
df.describe()
```

## Langkah kedua yaitu tampilkan 10 record pertama dari dataset dengan attribute bertipe angka menggunakan select_dtypes(include=['tipe data'].head(10)

```python
df.select_dtypes(include=['float64']).head(10)
```

## Langkah ketiga yaitu tampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.tentunya menggunakan prettytable dan dictkeys.

```python
dt = {}

dt['Jumlah Record'] = df.shape[0]
dt['Jumlah Attribute'] = df.shape[1]
dt['Nama Attribute'] = df.columns.tolist()
dt['Jumlah Attribute Dengan Tipe Angka'] = df.select_dtypes(include=['float64']).shape[1]

dictKeys = list(dt.keys())

table = prty.PrettyTable()
table.add_column(dictKeys[0], [dt['Jumlah Record']])
table.add_column(dictKeys[1], [dt['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(dt['Nama Attribute'])])
table.add_column(dictKeys[3], [dt['Jumlah Attribute Dengan Tipe Angka']])

print(table)
```
