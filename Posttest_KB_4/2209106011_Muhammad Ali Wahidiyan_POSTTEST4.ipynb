## Lakukan Data cleaning pada data dengan nilai null (jika tidak ada nilai null pada dataset, maka buatlah menjadi ada). Ganti nilai null sesuai ketentuan. (bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus),untuk itu di sini hanya mempunyai 2 tipe yaitu float dan object.

```python
df_4= df.copy()
df_4['MinTemp'].fillna(df_4['MinTemp'].mean(),inplace=True)
df_4['MaxTemp'].fillna(df_4['MinTemp'].mean(),inplace=True)
df_4['Evaporation'].fillna(df_4['Evaporation'].mean(),inplace=True)
df_4['Rainfall'].fillna(df_4['Rainfall'].mean(),inplace=True)
df_4['Sunshine'].fillna(df_4['Sunshine'].mean(),inplace=True)
df_4['WindGustSpeed'].fillna(df_4['WindGustSpeed'].mean(),inplace=True)
df_4['WindSpeed9am'].fillna(df_4['WindSpeed9am'].mean(),inplace=True)
df_4['WindSpeed3pm'].fillna(df_4['WindSpeed3pm'].mean(),inplace=True)
df_4['Humidity9am'].fillna(df_4['Humidity9am'].mean(),inplace=True)
df_4['Humidity3pm'].fillna(df_4['Humidity3pm'].mean(),inplace=True)
df_4['Pressure9am'].fillna(df_4['Pressure9am'].mean(),inplace=True)
df_4['Pressure3pm'].fillna(df_4['Pressure3pm'].mean(),inplace=True)
df_4['Cloud9am'].fillna(df_4['Cloud9am'].mean(),inplace=True)
df_4['Cloud3pm'].fillna(df_4['Cloud3pm'].mean(),inplace=True)
df_4['Temp9am'].fillna(df_4['Temp9am'].mean(),inplace=True)
df_4['Temp3pm'].fillna(df_4['Temp3pm'].mean(),inplace=True)

df_4['WindGustDir'].fillna(df_4['WindGustDir'].mode()[0],inplace=True)
df_4['WindDir9am'].fillna(df_4['WindDir9am'].mode()[0],inplace=True)
df_4['WindDir3pm'].fillna(df_4['WindDir3pm'].mode()[0],inplace=True)
df_4['RainToday'].fillna(df_4['RainToday'].mode()[0],inplace=True)
df_4['RainTomorrow'].fillna(df_4['RainTomorrow'].mode()[0],inplace=True)
```
